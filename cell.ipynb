```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
# Dropped "ıd" and "Unnamed: 32" columns.
data.drop(["id","Unnamed: 32"],axis=1, inplace=True)
```

```python
data.head()
```

```python
M = data[data["diagnosis"] == "M"]
B = data[data["diagnosis"] == "B"]
```

```python
# Scatter Plot
plt.scatter(M["texture_mean"],M["radius_mean"], color="red",label = "BAD")
plt.scatter(B["texture_mean"],B["radius_mean"], color="green",label = "GOOD")
plt.xlabel("Texture Mean")
plt.ylabel("Radius Mean")
plt.legend()
plt.show()
```

```python
data.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()

# M = 1, B = 0
data["diagnosis"] = label_encode.fit_transform(data["diagnosis"])
```

```python
data["diagnosis"].value_counts()
```

```python
data.head()
```

```python
# Selected x and y variable
y = data["diagnosis"]
x = data.drop("diagnosis",axis=1)
```

```python
data.head(3)
```

```python
# Normalize the Numerical Features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x = scaler.fit_transform(x)
x = pd.DataFrame(x, columns=data.drop("diagnosis",axis=1).columns)
x.head()

```

```python
# Train and Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.3)
```

```python
# KNN Model
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=3) # k = 3
knn_model = knn_classifier.fit(x_train, y_train)

# KNN Model Score
accuracy = knn_model.score(x_test, y_test) 
print(f"Accuracy : {accuracy}")
```

```python
# KNN Prediction and Accuracy Score
from sklearn.metrics import accuracy_score
y_pred = knn_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Model accuracy:", accuracy)
```

```python
# Actual and Prediction Value
result = pd.DataFrame({"Actual":y_test,"Prediction":y_pred})
result.head()
```

```python
# Find best K-value
all_scores = []

best_k = 1
best_score = 0

for k_value in range(1,15):
    knn_classifier_2 = KNeighborsClassifier(n_neighbors= k_value) 
    knn_classifier_2.fit(x_train, y_train)
    scores = knn_classifier_2.score(x_test, y_test)
    
    print("Scores:",scores," K-values :", k_value)

    if scores > best_score:
        best_score = scores
        best_k = k_value

print("**************")
print(f"Best Accuracy: {best_score}")
print(f"Best K-Value {best_k}")


```
